# Intro to LLMs using PydanticAI and Openrouter API

```python
from pydantic_ai import Agent
from dotenv import load_dotenv

load_dotenv()

agent = Agent(
    "openrouter:openrouter/healer-alpha",
    system_prompt=(
        "You are a nerdy joking programming bot that will always answer with a programming joke",
        "Answer with 3-5 sentences"
    ),
)

result = await agent.run("why is it foggy in göteborg")

result
```

## Basic prompt engineering

```python
chef_bot = Agent(
    "openrouter:meta-llama/llama-3.3-70b-instruct:free",
    system_prompt=(
        "You are a professional chef that is very good at making asian food and explaining in a pedagogical way how to create yummy dishes.",
        """
        Answer in the following format:
        
        Description 1-2 sentences of what the dish is
        
        List of ingredients in bullet points

        Short step by step description on how to make the dish
        """,
    ),
)

result = await chef_bot.run("Give me 3 asian soups in a list")
result.output
```

## Temperature


```python
from pydantic_ai import ModelSettings

agent = Agent(
    "openrouter:arcee-ai/trinity-mini:free",
    system_prompt="You are a children storyteller. Answer max 3 sentences.",
    model_settings=ModelSettings(temperature=0),
)

prompt = "A gray netherland rabbit jumps ..."

result = await agent.run(prompt)
print(result.output)
```

```python
result = await agent.run(prompt)
print(result.output)
```

```python
agent.model_settings = ModelSettings(temperature=2)
result = await agent.run(prompt)
print(result.output)
```

## Multimodal inputs

```python
from pathlib import Path
from pydantic_ai.messages import BinaryContent

image_data = Path("kokchun.png").read_bytes()
prompt = "What can you tell me about this dude"

image_joker = Agent(
    "openrouter:openrouter/healer-alpha",
    system_prompt="You are a person that can analyze and image, explain it in a funny way and max 3 sentences",
)

result = await image_joker.run(
    user_prompt= [prompt, BinaryContent(image_data, media_type="image/png")]
)

print(result.output)
```

```python
print(result.response.parts[0].content)
```
